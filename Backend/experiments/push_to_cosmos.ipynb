```python
from azure.cosmos import CosmosClient, exceptions
from typing import Dict, Any
import uuid


def push_json_to_cosmos(
    *,
    endpoint: str,
    key: str,
    database_name: str,
    container_name: str,
    json_data: Dict[str, Any],
    partition_key_field: str = "id",
) -> Dict[str, Any]:
    """
    Push a JSON object into a Cosmos DB container.

    Args:
        endpoint: Cosmos DB account endpoint
        key: Cosmos DB account key
        database_name: Database name
        container_name: Container name
        json_data: JSON object to insert
        partition_key_field: Field used as partition key

    Returns:
        Inserted document
    """

    try:
        # Create Cosmos client
        client = CosmosClient(endpoint, credential=key)

        # Get database and container
        database = client.get_database_client(database_name)
        container = database.get_container_client(container_name)

        # Ensure document has an id
        if "id" not in json_data:
            json_data["id"] = str(uuid.uuid4())

        # Ensure partition key exists
        if partition_key_field not in json_data:
            raise ValueError(
                f"Partition key field '{partition_key_field}' missing in json_data"
            )

        # Insert document
        response = container.create_item(body=json_data)

        print(f"Document inserted successfully with id: {response['id']}")

        return response

    except exceptions.CosmosHttpResponseError as e:
        print(f"Cosmos DB Error: {e}")
        raise

    except Exception as e:
        print(f"Unexpected Error: {e}")
        raise
```
